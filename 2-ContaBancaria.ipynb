```python
from datetime import datetime # usei esse código para obter a data atual, para ser usada na contaPoupança

class ContaBancaria: # criei a classe e abaixo criei as propriedades solicitadas no exercicio 2
    def __init__(self, nome_cliente, num_conta, agencia, saldo):
        self.nome_cliente = nome_cliente
        self.num_conta = num_conta
        self.agencia = agencia
        self.saldo = saldo

    def sacar(self, valor): #CRIANDO condição para ser permitido somente SALDO POSITIVO
        if valor > self.saldo:
            print("Saldo insuficiente para saque.")
        else:
            self.saldo -= valor

    def depositar(self, valor):
        if valor > 0:
            self.saldo += valor
        else:
            print("Valor do depósito deve ser positivo.")

    def exibir_dados(self): # criado função para retornar as propriedades listadas acima
        return (f"Nome Cliente: {self.nome_cliente}\n"
                f"Número da Conta: {self.num_conta}\n"
                f"Agência: {self.agencia}\n"
                f"Saldo: R${self.saldo:.2f}\n")


# Classe HERDADA da ContaBancaria
class ContaPoupanca(ContaBancaria):
    def __init__(self, nome_cliente, num_conta, agencia, saldo, dia_rendimento, taxa_rendimento):
        super().__init__(nome_cliente, num_conta, agencia, saldo)
        self.dia_rendimento = dia_rendimento
        self.taxa_rendimento = taxa_rendimento

    def calcular_novo_saldo(self):
        hoje = datetime.now().day
        if hoje == self.dia_rendimento:
            self.saldo += self.saldo * (self.taxa_rendimento / 100)

    def exibir_dados(self):
        return (super().exibir_dados() +
                f"Dia de Rendimento: {self.dia_rendimento}\n"
                f"Taxa de Rendimento: {self.taxa_rendimento}%\n")


# Classe HERDADA da ContaBancaria
class ContaEspecial(ContaBancaria):
    def __init__(self, nome_cliente, num_conta, agencia, saldo, valor_limite):
        super().__init__(nome_cliente, num_conta, agencia, saldo)
        self.valor_limite = valor_limite

    def sacar(self, valor):
        if valor > (self.saldo + self.valor_limite):
            print("Saldo e limite insuficientes para saque.")
        else:
            self.saldo -= valor

    def exibir_dados(self):
        return (super().exibir_dados() +
                f"Valor Limite: R${self.valor_limite:.2f}\n")


# Instanciando e testando os objetos
if __name__ == "__main__":
    # Criando e testando ContaBancaria
    conta1 = ContaBancaria("SIMONE MAYARA", "56148", "007", 6000.00)
    conta1.depositar(3580.00)
    conta1.sacar(100.00)
    print("Conta Bancária:")
    print(conta1.exibir_dados())

    # Criando e testando a ContaPoupanca
    conta2 = ContaPoupanca("MAYARA SOARES", "98470", "002", 3500.00, 4, 3.5)
    conta2.depositar(2400.00)
    conta2.sacar(99.00)
    conta2.calcular_novo_saldo() #TESTANDO que hoje é o dia de rendimento
    print("Conta Poupança:")
    print(conta2.exibir_dados())

    # Criando e testando a ContaEspecial
    conta3 = ContaEspecial("HEBER VALERIANO", "49821", "008", 500.00, 1000.00)
    conta3.depositar(300.00)
    conta3.sacar(1450.00)  #TESTANDO saque dentro do limite, TESTANDO CONTA NEGATIVA
    print("Conta Especial:")
    print(conta3.exibir_dados())
```

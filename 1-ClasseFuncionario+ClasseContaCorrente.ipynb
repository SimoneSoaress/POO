```python
from datetime import datetime #IMPORTEI DATETIME PARA MANIPULAÇÃO DE DATA E HORA.

class Funcionario:
    def __init__(self, matricula, nome, data_admissao, valorHora, valorHoraExtra, cargo, status): #USO DEF PARA DEFINIR UMA FUNÇÃO, UMA VEZ DEFINIDA POSSO CHAMA-LÁ VÁRIAS VEZES
        self.matricula = matricula  #SELF REFERE-SE AO OBJETO
        self.nome = nome
        self.data_admissao = datetime.strptime(data_admissao, "%d-%m-%Y") #AQUI USEI O DATETIME.STRPTIME PARA CONVERTER STRING DATA PARA OBJETO e MANIPULAR A EXIBIÇÃO DO FORMATO.
        self.valorHora = valorHora
        self.valorHoraExtra = valorHoraExtra
        self.cargo = cargo
        self.status = status

    def calcularSalario(self, qtde_horas_trabalhadas): # AQUI CRIEI UMA FUNÇÃO ONDE CHAMO RETURN PARA CALCULAR O VALOR DA HORA VEZES A QUANTIDADE DE HORAS TRABALHADAS.
        return self.valorHora * qtde_horas_trabalhadas

    def calcular_total_horas_extras(self, qtde_horas_extras_trabalhadas): #FUNÇÃO PARA CALCULAR HORAS EXTRAS
        return self.valorHoraExtra * qtde_horas_extras_trabalhadas

    def reajustar_valor_hora(self, percentual):
        self.valorHora += self.valorHora * (percentual / 100) # VALOR ATUAL DA HORA TRABALHADA DO FUNCIONÁRIO | CONVERTE O PERCENTUAL PARA UMA FRAÇÃO DECIMAL.
        self.valorHoraExtra += self.valorHoraExtra * (percentual / 100) # CALCULA O VALOR DO REAJUSTE

# INSTANCIANDO UM OBJETO DA CLASSE FUNCIONARIO E TESTANDO SEUS RESPECTIVOS MÉTODOS.
funcionario = Funcionario(
    matricula= "039.2.23.036",
    nome= "SIMONE MAYARA",
    data_admissao="24-07-2024",
    valorHora=60,
    valorHoraExtra=85,
    cargo= "Analista",
    status= "Ativo"
)

# TESTANDO . . .
qtde_horas_trabalhadas = 160
qtde_horas_extras_trabalhadas = 24
percentual_reajuste = 10

salario = funcionario.calcularSalario(qtde_horas_trabalhadas)
total_horas_extras = funcionario.calcular_total_horas_extras(qtde_horas_extras_trabalhadas)
funcionario.reajustar_valor_hora(percentual_reajuste)

# PRINTS
print(f"Matrícula: {funcionario.matricula}")
print(f"Nome: {funcionario.nome}")
print(f"Data de Admissão: {funcionario.data_admissao.strftime('%d/%m/%Y')}")
print(f"Valor Hora: R${funcionario.valorHora}")
print(f"Valor Hora Extra: R${funcionario.valorHoraExtra}")
print(f"Salário: R${salario:,.2f}") # formatei para que a exibição do salário fosse separado por virgula na casa do milhar e . após duas casas decimais.
print(f"Total Horas Extras: {total_horas_extras:,.2f}")
print(f"Cargo: {funcionario.cargo}")
print(f"Status: {funcionario.status}")

# -------------------------------------------------------------------------------------------------------------------------------

class ContaCorrente: # criando a classe ContaCorrente
    def __init__(self, numeroConta, titular, saldo=0.0, limite=0.0): # o método __init__ é o construtor da classe. Ele é chamado quando um novo objeto da classe é criado.
        self.numeroConta = numeroConta #logo foi criado os paramêtros
        self.titular = titular
        self.saldo = saldo
        self.limite = limite

    def depositar(self, valor): #criando a função depositar e verificando se o valor depositado é positivo, antes de atualizar o SALDO.
        if valor > 0:
            self.saldo += valor
            return True
        return False

    def sacar(self, valor): # dentro da funçãoo sacar verifico se o valor do saque é positivo e se há saldo suficiente incluindo o limite.
        if valor > 0 and valor <= (self.saldo + self.limite):
            self.saldo -= valor
            return True
        return False

    def consultar_saldo(self):
        return self.saldo # exibe o saldo da conta

    def consultar_limite(self):
        return self.limite #exibe o limite de crédito da conta.


conta = ContaCorrente(numeroConta="039.2.23.036", # INSTANCIANDO.. OBJETO DA CLASSE CONTA CORRENTE E TESTANDO OS MÉTODOS
                      titular="MAYARA SOARES",
                      saldo=9800.0,
                      limite=650.0)

# TESTANDO . . .
deposito_sucesso = conta.depositar(200.0)
saque_sucesso = conta.sacar(1600.0)

# PRINTS
print("\n\n")
print(f"Número da Conta: {conta.numeroConta}")
print(f"Titular da Conta: {conta.titular}")
print(f"Saldo: R${conta.consultar_saldo()}")
print(f"Limite: R${conta.consultar_limite()}")
print(f"Depósito realizado: {'Sim' if deposito_sucesso else 'Não'}")
print(f"Saque realizado: {'Sim' if saque_sucesso else 'Não'}")
```
